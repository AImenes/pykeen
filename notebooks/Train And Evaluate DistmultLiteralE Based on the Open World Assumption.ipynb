```python
import logging
import os
import sys
import time
import poem
from poem.constants import KG_EMBEDDING_MODEL_NAME, DISTMULT_LITERAL_NAME_OWA, TRAINING_SET_PATH, PATH_TO_NUMERIC_LITERALS,KG_ASSUMPTION,OWA,EMBEDDING_DIM,MARGIN_LOSS,OPTMIZER_NAME,SGD_OPTIMIZER_NAME, LEARNING_RATE, NUM_EPOCHS, BATCH_SIZE, EVALUATOR,RANK_BASED_EVALUATOR,SEED, NUM_ENTITIES, NUM_RELATIONS
from poem.pipeline import Helper
import numpy as np
```

```python
logging.basicConfig(level=logging.INFO)
logging.getLogger('poem').setLevel(logging.INFO)
```

Define datapaths:

```python
literals_path = '../tests/resources/numerical_literals.txt'
triples_path = '../tests/resources/test.txt'
```

Define config:

```python
config = {
        KG_EMBEDDING_MODEL_NAME: DISTMULT_LITERAL_NAME_OWA,
        TRAINING_SET_PATH: triples_path,
        PATH_TO_NUMERIC_LITERALS: literals_path,
        KG_ASSUMPTION: OWA,
        EMBEDDING_DIM: 100,
        MARGIN_LOSS: 1,
        OPTMIZER_NAME: SGD_OPTIMIZER_NAME,
        LEARNING_RATE: 0.01,
        NUM_EPOCHS: 1,
        BATCH_SIZE: 128,
        EVALUATOR: RANK_BASED_EVALUATOR,
        SEED: 2

    }
```

Create instances:

```python
instances = Helper.preprocess_train_triples(config=config)
config[NUM_ENTITIES] = len(instances.entity_to_id)
config[NUM_RELATIONS] = len(instances.relation_to_id)
```

Create model config:

```python
model_config = Helper.create_model_config(config=config, instances=instances)
```

Configure KGE model:

```python
kge_model = Helper.get_kge_model(model_config=model_config)
```

Configure training loop:

```python
all_entities = np.array(list(instances.entity_to_id.values()))
```

```python
training_loop = Helper.get_training_loop(config=config, kge_model=kge_model, all_entities=all_entities)
```

Train KGE model:

```python
fitted_kge_model, losses = training_loop.train(training_instances=instances)
```

Get evaluator: 

```python
evaluator = Helper.get_evaluator(kge_model=fitted_kge_model,
                                 config=config,
                                 entity_to_id=instances.entity_to_id,
                                 relation_to_id=instances.relation_to_id,
                                 training_triples=instances.instances)
```

Evaluate KGE model:

```python
metric_results = evaluator.evaluate(test_triples=instances.instances[0:100])
```
